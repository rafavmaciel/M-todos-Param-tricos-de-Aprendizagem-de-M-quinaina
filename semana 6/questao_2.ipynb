```python
#carregando o dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('Skin_NonSkin.txt' , sep = '\t')

data.columns = ['x','y','z','classe']
data.head()
```

```python
data.describe()
```

```python
#gerando grfico de compartimentos hexagonal entre x e y
plt.figure(figsize=(10,10))
grafico = plt.hexbin(data.x, data.y, cmap='viridis' , gridsize=50)
plt.colorbar(grafico)
plt.xlabel('x' , fontsize=15)
plt.ylabel('y' , fontsize=15)
plt.title('Hexagonal bin plot')
plt.show()
```

```python
#gerando grfico de compartimentos hexagonal entre x e z
plt.figure(figsize=(10,10))
grafico = plt.hexbin(data.x, data.z, cmap='viridis',gridsize=50 )
plt.colorbar(grafico)
plt.xlabel('x' , fontsize=15)
plt.ylabel('z' , fontsize=15)
plt.title('Hexagonal bin plot')
plt.show()
```

```python
#gerando grfico de compartimentos hexagonal entre y e z

plt.figure(figsize=(10,10))
grafico = plt.hexbin(data.y, data.z, cmap='viridis', gridsize=50)
plt.colorbar(grafico)
plt.xlabel('y' , fontsize=15)
plt.ylabel('z' , fontsize=15)
plt.title('Hexagonal bin plot')
```

```python
#dividindo o dataset por classes
skin = data[data.classe == 1]
non_skin = data[data.classe == 2]

```

```python
#gerando grfico de compartimentos hexagonal entre x e y para a classe skin
plt.figure(figsize=(10,10))
grafico = plt.hexbin(skin.x, skin.y, cmap='viridis', gridsize=50)
plt.colorbar(grafico)
plt.xlabel('x' , fontsize=15)
plt.ylabel('y' , fontsize=15)
plt.title('Hexagonal da skin')
plt.show()
```

```python
#gerando grfico de compartimentos hexagonal entre x e z para a classe skin
plt.figure(figsize=(10,10))
grafico = plt.hexbin(skin.x, skin.z, cmap='viridis', gridsize=50)
plt.colorbar(grafico)
plt.xlabel('x' , fontsize=15)
plt.ylabel('z' , fontsize=15)
plt.title('Hexagonal da skin')
plt.show()
```

```python
#gerando grfico de compartimentos hexagonal entre y e z para a classe skin
plt.figure(figsize=(10,10))
grafico = plt.hexbin(skin.y, skin.z, cmap='viridis', gridsize=50)
plt.colorbar(grafico)
plt.xlabel('y' , fontsize=15)
plt.ylabel('z' , fontsize=15)
plt.title('Hexagonal da skin')
```

```python
#gerando grfico de compartimentos hexagonal entre x e y para a classe non skin
plt.figure(figsize=(10,10))
grafico = plt.hexbin(non_skin.x, non_skin.y, cmap='viridis', gridsize=50)
plt.colorbar(grafico)
plt.xlabel('x' , fontsize=15)
plt.ylabel('y' , fontsize=15)
plt.title('Hexagonal da non skin')
plt.show()
```

```python
#gerando grfico de compartimentos hexagonal entrre x e z para a classe non skin
plt.figure(figsize=(10,10))
grafico = plt.hexbin(non_skin.x, non_skin.z, cmap='viridis', gridsize=50)
plt.colorbar(grafico)
plt.xlabel('x' , fontsize=15)
plt.ylabel('z' , fontsize=15)
plt.title('Hexagonal da non skin')
plt.show()
```
